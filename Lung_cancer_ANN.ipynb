```python
df=df.drop(['Name','Surname'],axis=1)
df
```

```python
df.isna().sum()
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
x=x.values
x
```

```python
y=y.values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
scale.fit(x_train)
x_train=scale.fit_transform(x_train)
x_test=scale.fit_transform(x_test)
```

```python
x_train
```

```python
x_test
```

Model creation

```python
#initialization

ann=tf.keras.models.Sequential()
```

```python
#first hidden layer

ann.add(tf.keras.layers.Dense(6,activation='relu'))
```

```python
#second hidden layer

ann.add(tf.keras.layers.Dense(6,activation='relu'))
```

```python
#add output layer

ann.add(tf.keras.layers.Dense(1,activation='sigmoid'))
```

```python
# compiling artificial neural network(ANN)

ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
#fit

ann.fit(x_train,y_train,batch_size=32,epochs=100)
```

```python
x_train.shape
```

```python
pred = ann.predict([[10,0,5,2]])
print(pred)
```

```python
# apply normalization apply

pred = ann.predict(scale.transform([[30,0,5,2]]))
if(pred>0.5):
  print(1)
else:
  print(0)

```
